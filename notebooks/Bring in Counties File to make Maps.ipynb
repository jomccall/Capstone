```python
import addfips
import requests
import matplotlib.pyplot as plt
from IPython.display import Image
import folium
import pickle
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
import pyproj
import collections
import numpy as np

%matplotlib inline
```

```python
counties_geo = gpd.read_file('../data/test/countybreakout.shp')
```

```python
counties_geo['global_rare'] = counties_geo['G1'] + counties_geo['G12'] + counties_geo['G2'] + counties_geo['G3']
```

```python
counties_geo['global_notrare'] = counties_geo['G34'] + counties_geo['G4'] + counties_geo['G45'] + counties_geo['G5'] 
```

GRank 1

```python
fig, ax = plt.subplots(figsize=(20, 6))
counties_geo.plot(column = 'G1',
               cmap = 'Greens',
               edgecolor = 'grey',
               legend = True,
               ax = ax)
for index, row in counties_geo.iterrows():
    plt.annotate(text=row['G1'], 
                 xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                 horizontalalignment='center')
plt.title('Density of G1 Species Observations', fontsize = 16, fontweight = 'bold')
ax.axis('off');
```
